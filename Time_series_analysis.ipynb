```python
df['USD (PM)'].fillna(method='ffill',inplace=True)
df['GBP (PM)'].fillna(method='ffill',inplace=True)
df['EURO (PM)'].fillna(method='ffill',inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['day']=pd.to_datetime(df['Date']).dt.day
df['month']=pd.to_datetime(df['Date']).dt.month
df
```

```python
df=df.drop(['Date'],axis=1)
df
```

```python
for i in range(6,13):
  df['month']==i
  df_mean=df[df['month']==i].mean()
  print(df_mean)

```

```python
for i in range(6,13):
  df['month']==i
  print('month:',i)
  fig1 = px.line(df[df['month']==i],x='day',y='USD (PM)',markers=True)
  fig1.show()
```

```python
df['USD (PM)'].sort_values(ascending=False)
```
